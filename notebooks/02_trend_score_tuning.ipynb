# 🎛️ ONYX Trend Score Tuning Sandbox
Adjust weights and refine the market trend detection algorithm.

```python
import math
import pandas as pd

def simulate_trend_score(rank, rating, review_count, weights):
    """
    Weights dict: {'rank': 0.4, 'rating': 0.2, 'reviews': 0.4}
    """
    w_rank = weights.get('rank', 0.4)
    w_rating = weights.get('rating', 0.2)
    w_reviews = weights.get('reviews', 0.4)
    
    # Normalization logic
    norm_rank = 1.0 / rank if rank > 0 else 0
    norm_reviews = math.log10(review_count) if review_count > 1 else 0
    
    score = (w_rank * norm_rank) + (w_rating * rating) + (w_reviews * (norm_reviews / 7))
    return round(score, 4)
```

### 1. Test Scenarios
Compare how different games score under your current weights.

```python
current_weights = {'rank': 0.5, 'rating': 0.3, 'reviews': 0.2}

scenarios = [
    {'title': 'Viral Indie Hit', 'rank': 1, 'rating': 4.8, 'reviews': 10000},
    {'title': 'Established AAA', 'rank': 10, 'rating': 4.2, 'reviews': 1000000},
    {'title': 'Falling Star', 'rank': 100, 'rating': 3.5, 'reviews': 50000}
]

for s in scenarios:
    score = simulate_trend_score(s['rank'], s['rating'], s['reviews'], current_weights)
    print(f"{s['title']}: {score}")
```

### 2. Strategy Comparison
What if we prioritize 'Rating' over 'Rank'?

```python
strategy_a = {'rank': 0.7, 'rating': 0.2, 'reviews': 0.1} # High Rank Velocity
strategy_b = {'rank': 0.2, 'rating': 0.7, 'reviews': 0.1} # Quality First

for s in scenarios:
    score_a = simulate_trend_score(s['rank'], s['rating'], s['reviews'], strategy_a)
    score_b = simulate_trend_score(s['rank'], s['rating'], s['reviews'], strategy_b)
    print(f"{s['title']} -> Strategy A: {score_a} | Strategy B: {score_b}")
```
