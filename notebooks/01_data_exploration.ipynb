# 🧪 ONYX Market Data Exploration
Use this notebook to explore the raw market snapshots stored in the SQLite database.

```python
import pandas as pd
import sqlite3
import plotly.express as px
import os

# Connect to the source of truth
db_path = "../database/market_analyzer.db"
conn = sqlite3.connect(db_path)

print(f"Connected to: {os.path.abspath(db_path)}")
```

### 1. Snapshot Summary
Let's see how much data we've collected across each category.

```python
query = """
SELECT 
    category, 
    platform, 
    COUNT(*) as item_count, 
    DATE(fetched_at) as fetch_date
FROM snapshots 
GROUP BY category, platform, fetch_date
"""
df_summary = pd.read_sql(query, conn)
df_summary
```

### 2. Genre Distribution
Which genres are dominating the market snapshots currently?

```python
df_all = pd.read_sql("SELECT * FROM snapshots", conn)
fig = px.pie(df_all, names='genre', title='Global Genre Distribution (All Snapshots)', hole=0.3)
fig.show()
```

### 3. Rating vs. Review Count
Are the most reviewed games also the highest rated?

```python
fig = px.scatter(df_all, x='reviews', y='rating', color='category', 
                 hover_name='title', log_x=True, 
                 title='Rating vs. Review Count (Log Scale)')
fig.show()
```
